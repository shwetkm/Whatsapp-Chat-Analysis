## Reading file & pre-processing

```python
import pandas as pd
```

```python
f = 'D:/Personal Docs/mishragang.csv'
```

```python
list_ = []
with open(f,encoding="utf8") as mg:
    for line in mg:
        try:
            if(type(int(line[0:2])) == int):
               list_.append (line)
        except:
            pass
```

```python
mg = pd.DataFrame([sub.split(",") for sub in list_])
```

```python
mg.columns = ['Date','Time','Sender','Content1','Content2','Content3','Content4','Content5','Content6','Content7','Content8','Content9','Content10','Content11','Content12']
```

```python
mg.Time = mg.Time.apply(lambda x: (x.strip())[0:2])
```

```python
mg.head()
```

## Top texters of Mishra Gang

```python
sendr = pd.DataFrame(mg.Sender.value_counts())
sendr.reset_index(level=0, inplace=True)
sendr.columns = ['Sender','Count']
sendr.head(15)
```

```python
from bokeh.charts import Bar, output_file, show
from bokeh.io import output_notebook
output_notebook()
from bokeh.models import HoverTool
% matplotlib inline

p = Bar(sendr.head(10), label='Sender', values='Count', agg='max',
        title="Number of texts by sender",color = 'Sender',legend=False)

show(p)
```

# Hottest days in Mishra Gang 
(Days with high message count)

```python
dat = pd.DataFrame(mg.Date.value_counts())
dat.reset_index(level=0, inplace=True)
dat.columns = ['Date','Count']
dat.head(15)
```

```python
p = Bar(dat.head(10), label='Date', values='Count', agg='max',
        title="Dates with large message count",color = 'Date',legend=False)

show(p)
```

# Most happening time of the Day

```python
tim = pd.DataFrame(mg.Time.value_counts())
tim.reset_index(level=0, inplace=True)
tim.columns = ['Time','Count']
tim.head(15)
```

```python
p = Bar(tim.head(10), label='Time', values='Count', agg='max',
        title="Most happening time of the day",color = 'Time',legend=False)

show(p)
```
